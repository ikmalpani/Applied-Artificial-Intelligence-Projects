```python
import pandas as pd

train_df=pd.read_json('train.json').sort_values(by="listing_id")
test_df=pd.read_json('test.json').sort_values(by="listing_id")

train_df.describe()
```

```python
test_df.describe()
```

```python
#using number of photos as a metric instead of photos due to lack of computational power

train_df["num_photos"] = train_df["photos"].apply(len)

#creating new features based on hunches

train_df["num_features"] = train_df["features"].apply(len)
train_df["room_sum"] = train_df["bedrooms"]+train_df["bathrooms"]
#converting "created" column to change from object type to timeseries type
train_df[['created']] = pd.to_datetime(train_df['created'])
#Converting the target variables from low, medium and high to 1,2 and 3 for convinience
train_df = train_df.replace({"interest_level": {"high": 1, "medium": 2, "low": 3}})
#reseting the index to start from 0 for convinicence
train_df = train_df.reset_index()
train_df.head()
```

```python
#using number of photos as a metric instead of photos due to lack of computational power

test_df["num_photos"] = test_df["photos"].apply(len)

#creating new features based on hunches

test_df["num_features"] = test_df["features"].apply(len)
test_df["room_sum"] = test_df["bedrooms"]+test_df["bathrooms"]
#converting "created" column to change from object type to timeseries type
test_df[['created']] = pd.to_datetime(test_df['created'])
#Converting the target variables from low, medium and high to 1,2 and 3 for convinience
test_df = test_df.replace({"interest_level": {"high": 1, "medium": 2, "low": 3}})
#reseting the index to start from 0 for convinicence
test_df = test_df.reset_index()
test_df.head()
```

```python
from nltk.stem import PorterStemmer
import re

#Stemming for the description
stemmer = PorterStemmer()
def clean(x):
    regex = re.compile('[^a-zA-Z ]')
    i = regex.sub(' ', x).lower()
    i = i.split(" ") 
    i= [stemmer.stem(l) for l in i]
    i= " ".join([l.strip() for l in i if (len(l)>2) ]) # Keeping words that have length greater than 2
    return i
print('Description before Stemming: \n',train_df[['description']].head())
train_df[['description']] = train_df.description.apply(lambda x: clean(x))
print('Description after Stemming: \n',train_df[['description']].head())
```

```python
print('Description before Stemming: \n',test_df[['description']].head())
test_df[['description']] = test_df.description.apply(lambda x: clean(x))
print('Description after Stemming: \n',test_df[['description']].head())
```

```python
#Prices wrt to the number of bedrooms
train_df["price_t"] =train_df["price"]/train_df["bedrooms"]
# count of words present in description column
train_df["num_description_words"] = train_df["description"].apply(lambda x: len(x.split(" ")))
```

```python
#Prices wrt to the number of bedrooms
test_df["price_t"] =test_df["price"]/test_df["bedrooms"]
# count of words present in description column
test_df["num_description_words"] = test_df["description"].apply(lambda x: len(x.split(" ")))
```

```python
numerical_features = ["bathrooms","bedrooms","latitude","longitude","price","num_photos","num_features","room_sum"]
categorical_features = ["display_address", "manager_id", "building_id", "street_address"]
train_df[categorical_features].head()
```

```python
train_df[numerical_features].head()
```

```python
test_df[categorical_features].head()
```

```python
test_df[numerical_features].head()
```

```python
from sklearn import preprocessing

for f in categorical_features:
        if train_df[f].dtype=='object':
            #print(f)
            lbl = preprocessing.LabelEncoder()
            lbl.fit(list(train_df[f].values) + list(test_df[f].values))
            train_df[f] = lbl.transform(list(train_df[f].values))
            test_df[f] = lbl.transform(list(test_df[f].values))
            numerical_features.append(f)
```

```python
train_X = train_df[numerical_features]
test_X = test_df[numerical_features]
train_y = train_df['interest_level']
#test_y = test_df['interest_level']
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(train_X, train_y)
```

```python
y_pred = clf.predict_proba(test_X)
y_pred
```

```python
labels2idx = {label: i for i, label in enumerate(clf.classes_)}
print(labels2idx)
```

```python
test_df
```

```python
sub = pd.DataFrame()
sub['listing_id'] = test_df['index']
i = 0
for label in [ "high", "medium", "low"]:
    i = i + 1
    sub[label] = y_pred[:, labels2idx[i]]
print(sub)
sub.to_csv("submission_rf.csv", index=False)
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=25)
scores = cross_val_score(clf, train_X, train_y, cv=100)
scores.max()
```

```python
from sklearn.ensemble import ExtraTreesClassifier

clf = ExtraTreesClassifier()
scores2 = cross_val_score(clf, train_X, train_y, cv=100)
scores2.max()
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
scores3 = cross_val_score(clf, train_X, train_y, cv=100)
scores3.max()
```

```python
from sklearn.metrics import log_loss
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train_X, train_y, test_size=0.20)

print("Random Forests:\nTraining...n_estimators=1000")
clf1 = RandomForestClassifier(n_estimators=1000)
clf1.fit(X_train, y_train)
y_val_pred = clf1.predict_proba(X_val)
print(log_loss(y_val, y_val_pred))
```

```python
print("Extra Trees:\nTraining...n_estimators=1000")
clf2 = ExtraTreesClassifier(n_estimators=1000)
clf2.fit(X_train, y_train)
y_val_pred1 = clf2.predict_proba(X_val)
print(log_loss(y_val, y_val_pred1))
```

```python
print("Adaboost:")
clf3 = AdaBoostClassifier()
clf3.fit(X_train, y_train)
y_val_pred2 = clf3.predict_proba(X_val)
print(log_loss(y_val, y_val_pred2))
```

```python
print("Logistic Regression:")
clf4 = LogisticRegression()
clf4.fit(X_train, y_train)
y_val_pred3 = clf4.predict_proba(X_val)
print(log_loss(y_val, y_val_pred3))
```

```python
from sklearn.ensemble import VotingClassifier

print("Voting Classifier:")
clf = VotingClassifier(estimators=[('rf',clf1),('lr',clf4)],voting='soft')
clf.fit(X_train, y_train)
y_val_pred4 = clf.predict_proba(X_val)
print(log_loss(y_val, y_val_pred4))
```
